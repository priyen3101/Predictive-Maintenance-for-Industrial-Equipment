### Handle imbalanced dataset & spliting in Train-Test

```python
df["Failure Type"].replace({"No Failure": 0,"Heat Dissipation Failure": 1,"Power Failure": 2,"Overstrain Failure": 3,"Tool Wear Failure": 4,"Random Failures": 5},inplace=True)
```

```python
df["Type"].replace({"H":0,"L":1,"M":2}, inplace=True)
```

```python
df.head(5)
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.33, stratify=y)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
y_train.value_counts()
```

```python
from imblearn.combine import SMOTETomek
smote=SMOTETomek(random_state=42)
x1_res,y1_res=smote.fit_resample(x_train,y_train)
```

```python
x1_res.head()
```

```python
y1_res.value_counts()
```

```python
df_x=pd.DataFrame(x1_res)
df_x['Type'].replace({0:"H",1:"L",2:"M"}, inplace=True)
```

```python
df_y=pd.DataFrame(y1_res)
df_y["Failure Type"].replace({0:"No Failure",1:"Heat Dissipation Failure",2:"Power Failure",3:"Overstrain Failure",4:"Tool Wear Failure",5:"Random Failures"},inplace=True)
```

```python
df1=pd.concat([df_x,df_y],axis=1)
```

```python
df1.head()
```

```python
_=sns.relplot(x='Tool wear [min]',y='Rotational speed [rpm]',hue='Failure Type',size='Failure Type',sizes=(60,5),data=df1)
```

### Preprocessing 

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import (OneHotEncoder,StandardScaler,PowerTransformer)
from sklearn.compose import ColumnTransformer
# Build categorical preprocessor

categorical_column=[col for col in df1.iloc[:,:-1].columns if df1[col].dtype=='O']
categorical_pipeline=make_pipeline(OneHotEncoder(sparse_output=False,handle_unknown="ignore"))

# Build numeric processor

to_log=['Rotational speed [rpm]','Tool wear [min]']
to_scal=['Air temperature [K]','Process temperature [K]','Torque [Nm]']
numeric_pipe_1=make_pipeline(PowerTransformer())
numeric_pipe_2=make_pipeline(StandardScaler())

# Full processor

full=ColumnTransformer(transformers=[('categorical',categorical_pipeline,categorical_column),('Power Transform',numeric_pipe_1,to_log),('Standrization',numeric_pipe_2,to_scal)])
```

### Random Forrest Classifier

```python
from sklearn.multiclass import OutputCodeClassifier

rand_forest=OutputCodeClassifier(RandomForestClassifier(),random_state=40,code_size=6)
```

```python
pipeline=Pipeline(steps=[('preprocess',full),('base',rand_forest)])
```

```python
model_random_forest=pipeline.fit(x1_res,y1_res)
y_pred_random_forest=model_random_forest.predict(x_test)
```

```python
from sklearn.metrics import f1_score,fbeta_score,matthews_corrcoef, log_loss, confusion_matrix,classification_report,make_scorer,balanced_accuracy_score
from sklearn.metrics import accuracy_score, roc_curve, auc,recall_score,roc_auc_score, average_precision_score, precision_score,precision_recall_curve,ConfusionMatrixDisplay, confusion_matrix
```

```python
print(classification_report(y_test,y_pred_random_forest))
print(confusion_matrix(y_test,y_pred_random_forest))
```

```python
fig, ax = plt.subplots(figsize=(12, 12))

# Create the matrix
cm = confusion_matrix(y_test, y_pred_random_forest)
cmp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=pipeline.classes_)
cmp.plot(ax=ax)
plt.show();
```

```python
print("Accuracy Score: %.4f" %(accuracy_score(y_test,y_pred_random_forest)))
print("Matthews correlation coefficient: %.4f" %(matthews_corrcoef(y_test,y_pred_random_forest)))
```

### Recommendation for Predictive Maintenance
#### CatBoost (especially One Vs Rest variant) stands out as the most effective model for predictive maintenance. It offers:

    High Accuracy (0.9809): Indicates strong performance across all classes.
    
    High Matthews Correlation Coefficient (0.7755): Reflects a good balance between precision and recall, essential for minimizing both FP and FN.
    
    Low False Positives and False Negatives: Particularly in the One Vs Rest variant, it shows very low FN and FP, which is critical for effective maintenance predictions.
